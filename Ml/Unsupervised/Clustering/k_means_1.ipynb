# *K-means (Scikit-Implementation)*

* *It uses concept of vector quantization(VQ).*

* *It works by dividing a large set of points (vectors) into groups having approximately the same number of points closest to them. Each group is represented by its centroid point.*

```python
import numpy as np
import pandas as pd
from sklearn import model_selection , preprocessing , cluster
from sklearn.cluster import KMeans
%matplotlib inline 
from matplotlib import style
style.use('ggplot')
import matplotlib.pyplot as plt
```

```python
X = np.array([[1,2],[1.5,1.8],[5,8],[8,8],[1,0.6],[9,11]])
plt.scatter(X[:,0] , X[:,1] , s=150 , linewidths=2)
plt.show()
```

```python
clf = cluster.KMeans(n_clusters=2)
clf.fit(X)
centroids = clf.cluster_centers_
labels = clf.labels_ # array containing labels of features -> y

# print(centroids)
```

```python
colors = ['g.' , 'k.'  ,'c.' ,'b.' ,'r.' ,'y.']
for i in range(len(X)):
    plt.plot(X[i][0] , X[i][1] , colors[labels[i]] ,markersize=10)
plt.scatter(centroids[:,0]  ,centroids[:,1] , marker='x', s=20 ,linewidths=2 )
plt.show()
    
```
