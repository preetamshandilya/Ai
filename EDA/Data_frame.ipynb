```python
import pandas as pd
import Quandl
```

```python
api_key = "zbuPWr3CoQXjNje1Z2qX"

states = pd.read_html('https://www.yourdictionary.com/articles/state-abbreviations')
main_df = pd.DataFrame()
for abbv in states[0][1][1:]:
    query = "FMAC/HPI_" + str(abbv)
    df = Quandl.get(query, authtoken=api_key)
    
    # Drop 'NSA Value' column if it exists
    if 'NSA Value' in df.columns:
        df.drop(columns=['NSA Value'], inplace=True)
    
    # Rename 'SA Value' column to the abbreviation (abbv)
    if 'SA Value' in df.columns:
        df.rename(columns={'SA Value': abbv}, inplace=True)
    
    if main_df.empty:
        main_df = df
    else:
        main_df = main_df.join(df, lsuffix='_main', rsuffix='_' + str(abbv))

print(main_df.head())

```
